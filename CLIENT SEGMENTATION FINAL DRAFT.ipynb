```python
def regexExtractor(string):
    notAlphabeticRegex = re.compile(r'[^A-Za-z]')
    excludeNotAlphabetic = notAlphabeticRegex.findall(string)
    postAlphabeticRegex = re.sub(notAlphabeticRegex, ' ', string)
    postSpacesTreatment = re.sub('\s+', ' ', postAlphabeticRegex)
    
    return postSpacesTreatment
```

```python
wordCount = dict()
spellMistakeWordCount = dict()
for i in range(0,len(df)):
    #print(df.ix[i])
    try:
        df.ix[i] = regexExtractor(df.ix[i].lower())
        tokens = df.ix[i].split(' ')
        newToken = []
        for token in tokens:
            #if token == 'mgr':
                #print(df.ix[i])
            if ac.spell(token) == token:
                newToken.append(token)
                #if token in ac.word_lists.LOWERCASE + ac.word_lists.LOWERED:     
                if token in wordCount.keys():
                    wordCount[token] += 1
                else:
                    wordCount[token] = 1
            else:
                newToken.append(ac.spell(token))
                if ac.spell(token) in wordCount.keys():
                    wordCount[ac.spell(token)] += 1
                else:
                    wordCount[ac.spell(token)] = 1
                    
                if token in spellMistakeWordCount.keys():
                    continue
                else:
                    spellMistakeWordCount[token] = ac.spell(token)
        df.ix[i] = ' '.join([x for x in newToken])
        #print(df.ix[i])
    except:
        #print(df.ix[i])
        continue
```

```python
df = clientTitle['title']
example = pd.DataFrame()
example['title'] = df.copy()
example['transformed_title'] = example.title.copy()
```

```python
for i in range (0, len(example)):
    rank = float('Inf')
    x = 0
    for transformedPosition, posLists in rules.items():
        for key, value in posLists.items():
            if key == 'Name':
                for position in posLists[key]:
                    if type(position) == list:
                        boolean = [x in str(example.title.ix[i]).lower() for x in position]
                        if all(boolean):
                            x += 1
                            if rank > posLists['Rank']:
                                rank = posLists['Rank']
                                example.transformed_title.ix[i] = transformedPosition
                    else:
                        if position.lower() in str(example.title.ix[i]).lower():
                            x += 1
                            if rank > posLists['Rank']:
                                rank = posLists['Rank']
                                example.transformed_title.ix[i] = transformedPosition
             
            elif key == 'Abbr':
                for position in posLists[key]:
                    if position in str(example.title.ix[i]):
                        x += 1
                        if rank > posLists['Rank']:
                            rank = posLists['Rank']
                            example.transformed_title.ix[i] = transformedPosition
            
            if x == 0:
                example.transformed_title.ix[i] = 'other'
```
